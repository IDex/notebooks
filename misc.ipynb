```python
import scipy as sc
import scipy.stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
theta=sc.stats.poisson(20).rvs(100000)
sigma=sc.stats.poisson(30).rvs(100000)
p=sc.stats.beta(theta,sigma)
sns.distplot(p.rvs(100000))
plt.figure()
sns.distplot(theta)
```

```python
x=sc.linspace(0,1,100)
y=sc.stats.beta(20,30).pdf(x)
plt.plot(x,y)
```
